```python
def return_syndrome_hash_map(PARITY_CHECK_MATRIX, distance):
    SYNDROME_HASH_MAP = {}
    correctionInterval = math.floor((distance - 1) / 2)
    codedWordLength = (np.array(PARITY_CHECK_MATRIX).shape)[1]
    PARITY_CHECK_MATRIX_TRANSPOSE = PARITY_CHECK_MATRIX.getT()
    TEST_MESSAGE = [0 for i in range(codedWordLength)]
    INDEXES = [i for i in range(codedWordLength)]
    for bitflipsNum in range(correctionInterval + 1):
        combinations = list(itt.combinations(INDEXES, bitflipsNum))
        for bitflipScenario in range(math.comb(codedWordLength, bitflipsNum)):
            for i in combinations[bitflipScenario]:
                TEST_MESSAGE[i] = 1
            syndromeTuple = tuple(
                np.array(
                    np.apply_along_axis(
                        lambda x: x % 2,
                        1,
                        np.matmul(TEST_MESSAGE, PARITY_CHECK_MATRIX_TRANSPOSE),
                    )
                )[0]
            )
            SYNDROME_HASH_MAP.setdefault(syndromeTuple, combinations[bitflipScenario])
            TEST_MESSAGE = [0 for i in range(codedWordLength)]
    return SYNDROME_HASH_MAP


GOLAY_24_12_8_2_SYNDROME_HASH_MAP = return_syndrome_hash_map(
    PARITY_CHECK_MATRIX_GOLAY_24_12_8_2, 8
)
```

```python
def return_syndrome_tuple(SYNDROME_HASH_MAP, PARITY_CHECK_MATRIX, ENCODED_MESSAGE):
    return SYNDROME_HASH_MAP[
        tuple(
            np.array(
                np.apply_along_axis(
                    lambda x: x % 2,
                    1,
                    np.matmul(ENCODED_MESSAGE, PARITY_CHECK_MATRIX.getT()),
                )
            )[0]
        )
    ]

def encode_msg(GENERATIVE_MATRIX,MSG):
    MSG = np.matrix(MSG)
    MSG = np.matmul(MSG,GENERATIVE_MATRIX)
    return np.apply_along_axis(lambda x: x % 2, 1, MSG)
```

```python
def ENCODE_CHANNEL_MESSAGE(msg):
    blocks = [b1, b2, b3, b4, b5, b6, b7, b8, b9] = [
        msg[:12],
        msg[12:24],
        msg[24:36],
        msg[36:48],
        msg[48:60],
        msg[60:72],
        msg[72:84],
        msg[84:96],
        msg[96:],
    ]
    encodedMSG = []
    for block in blocks[:8]:
       encodedBlockMSG = encode_msg(GENERATIVE_MATRIX_GOLAY_24_12_8_2,block)
       encodedBlockMSG = list(np.array(encodedBlockMSG)[0])
       encodedMSG.extend(encodedBlockMSG)
    encodedBlockMSG = encode_msg(GENERATIVE_MATRIX_HAMMING_7_4_3_2,b9)
    encodedBlockMSG = list(np.array(encodedBlockMSG)[0])
    encodedMSG.extend(encodedBlockMSG)
    return encodedMSG
    
```

```python
# Postavljeno na ispravljanje
def DECODE_CHANNEL_MESSAGE(encodedMsg):
    blocks = [b1, b2, b3, b4, b5, b6, b7, b8, b9] = [
        encodedMsg[:24],
        encodedMsg[24:48],
        encodedMsg[48:72],
        encodedMsg[72:96],
        encodedMsg[96:120],
        encodedMsg[120:144],
        encodedMsg[144:168],
        encodedMsg[168:192],
        encodedMsg[192:],
    ]
    for offset, block in enumerate(blocks[:8]):
        synTuple = return_syndrome_tuple(
            GOLAY_24_12_8_2_SYNDROME_HASH_MAP,
            PARITY_CHECK_MATRIX_GOLAY_24_12_8_2,
            np.matrix(block),
        )
        for index in synTuple:
            encodedMsg[offset * 24 + index] = encodedMsg[offset * 24 + index] ^ 1
    synTuple = return_syndrome_tuple(
        HAMMING_7_4_3_2_SYNDROME_HASH_MAP,
        PARITY_CHECK_MATRIX_HAMMING_7_4_3_2,
        np.matrix(b9),
    )
    for index in synTuple:
        encodedMsg[192+ index] = encodedMsg[192+ index] ^ 1
    return encodedMsg
```
